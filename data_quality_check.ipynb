#### Data Quality issues and cleaning

#### Data formatting issues

##### > *  _id,createDate, dateScanned, finishedDate , modifyDate, pointsAwardedDate, purchaseDate, rewardsReceiptItemList are all JSON object literals that need to be converted to string literals for data cleaning (formatting issues)
##### > * All dates have been recorded as UNIX time format so they need to be converted to pandas datetime data type

#### Data inconsistency, structure, accesiility and readability issues
##### > * The column rewardsReceiptItemList can be seen as a list of dictionaries in every row so some relevant key,value pairs need to extracted out

##### > *  In the column rewardsReceiptItemList, different rows contain different number of dictionaries, containing different key-value pairs leading to data inconsistency (accessibility and readability issues)

#### Incomplete data
##### > *  A lot of columns and rows in this dataset contain null values, which leads to incomplete fields. Dropping these rows can lead to significant loss of information and data.

#### Data duplication and big size
##### > * Since, each receipt has multiple items and information about item list is in the field rewardsReceiptItemList, to get receipt level data,information would need to be extracted from each dictionary, exploding the data in size and also cause duplication at the receipt level (data storage structural problems and size issues)





```python
#since a lot of columns have the json structure in the columns, those columns that need cleaning up need to converted 
#from json literals to string literals for ease of data cleaning

df_receipts['_id'] = df_receipts['_id'].astype('str') 
df_receipts['createDate'] = df_receipts['createDate'].astype('str') 
df_receipts['dateScanned'] = df_receipts['dateScanned'].astype('str') 
df_receipts['finishedDate'] = df_receipts['finishedDate'].astype('str') 
df_receipts['modifyDate'] = df_receipts['modifyDate'].astype('str') 
df_receipts['pointsAwardedDate'] = df_receipts['pointsAwardedDate'].astype('str') 
df_receipts['purchaseDate'] = df_receipts['purchaseDate'].astype('str') 
```

```python
#cleaning the above columns using strip in built function that can be applied on strings
df_receipts['_id'] = df_receipts['_id'].str.strip('{}')
df_receipts['createDate']=df_receipts['createDate'].str.strip('{}')

df_receipts['dateScanned']=df_receipts['dateScanned'].str.strip('{}')
df_receipts['finishedDate']=df_receipts['finishedDate'].str.strip('{}')
df_receipts['modifyDate']=df_receipts['modifyDate'].str.strip('{}')
df_receipts['pointsAwardedDate']=df_receipts['pointsAwardedDate'].str.strip('{}')
df_receipts['purchaseDate']=df_receipts['purchaseDate'].str.strip('{}')

#further cleaning using string splicing
df_receipts['_id']=df_receipts['_id'].str[:-1]
df_receipts['createDate'] = df_receipts['createDate'].str[9:]

df_receipts['dateScanned'] = df_receipts['dateScanned'].str[9:]
df_receipts['finishedDate'] = df_receipts['finishedDate'].str[9:]
df_receipts['modifyDate'] = df_receipts['modifyDate'].str[9:]
df_receipts['pointsAwardedDate'] = df_receipts['pointsAwardedDate'].str[9:]
df_receipts['purchaseDate'] = df_receipts['purchaseDate'].str[9:]
df_receipts['_id']=df_receipts['_id'].str[9:]
```

```python
#changing the UNIX format of dates to pandas datetime readable format

df_receipts['createDate']=pd.to_datetime(df_receipts['createDate'], unit='ms')
df_receipts['dateScanned']=pd.to_datetime(df_receipts['dateScanned'], unit='ms')
df_receipts['finishedDate']=pd.to_datetime(df_receipts['finishedDate'], unit='ms')
df_receipts['modifyDate']=pd.to_datetime(df_receipts['modifyDate'], unit='ms')
df_receipts['purchaseDate']=pd.to_datetime(df_receipts['purchaseDate'], unit='ms')
df_receipts['pointsAwardedDate']=pd.to_datetime(df_receipts['pointsAwardedDate'], unit='ms')
```

```python
# on careful inspection of the rewardsReceiptItemList, explode function can be used to easily access receipt level properties 
#as well as extract key-value pairs that can help link this database to other databases
#for this explode() function is used to transform each element of a list-like to a row, replicating the index values. This is used on 
#rewardsReceiptItemList

df_receipts = df_receipts.explode('rewardsReceiptItemList')

#as mentioned before, the 'exploding' the rewardsReceiptItemList exponentially increases the data size from 1119 rows to 7381 rows (almost 7 times)
#it also introduces redundancy in data with duplicate receipt ids
```

```python
#resetting index for the dataset
df_receipts=df_receipts.reset_index()
```

```python
#extracting relevant key value pairs and putting them in a list especially for item barcode, userflaggedbarcode,
#item description and brand code to get better insights about the data, ease of readability and making it suitable to be 
#linked to other tables like brand dataset

list_barcodes=[]
list_desc=[]
list_user_barcode=[]

for i in range (0, len(df_receipts)):
    key1='barcode'
    key2='userFlaggedBarcode'
    key3='description'
    if key1 in str(df_receipts['rewardsReceiptItemList'][i]):
        x=df_receipts['rewardsReceiptItemList'][i]['barcode']
    else:
        x='Not Available'
    list_barcodes.append(x)
    if key2 in str(df_receipts['rewardsReceiptItemList'][i]):
        y=df_receipts['rewardsReceiptItemList'][i]['userFlaggedBarcode']
    else:
        y='Not Available'
    list_user_barcode.append(y)
    
    if key3 in str(df_receipts['rewardsReceiptItemList'][i]):
        z=df_receipts['rewardsReceiptItemList'][i]['description']
    else:
        z='Not Available'
    list_desc.append(z)
    
            
```

```python
list_brandcode=[]
for i in range (0, len(df_receipts)):
    key4='brandCode'
    if key4 in str(df_receipts['rewardsReceiptItemList'][i]):
        code=df_receipts['rewardsReceiptItemList'][i]['brandCode']
    else:
        code='Not Available'
    list_brandcode.append(code)
```

```python
reward_partner_id=[]
for i in range (0, len(df_receipts)):
    key6='rewardsProductPartnerId'
    if key6 in str(df_receipts['rewardsReceiptItemList'][i]):
        r_code=df_receipts['rewardsReceiptItemList'][i]['rewardsProductPartnerId']
    else:
        r_code='None'
    reward_partner_id.append(r_code)
```

```python
list_rewards_group=[]
for i in range (0, len(df_receipts)):
    key5='rewardsGroup'
    if key5 in str(df_receipts['rewardsReceiptItemList'][i]):
        code=df_receipts['rewardsReceiptItemList'][i]['rewardsGroup']
    else:
        code='None'
    list_rewards_group.append(code)
```

```python
points_payer_id=[]
for i in range (0, len(df_receipts)):
    key7='pointsPayerId'
    if key7 in str(df_receipts['rewardsReceiptItemList'][i]):
        pp_code=df_receipts['rewardsReceiptItemList'][i]['pointsPayerId']
    else:
        pp_code='None'
    points_payer_id.append(pp_code)
```

```python
## making a dataframe of the above lists

barcode_df = pd.DataFrame(
    {'Product_barcode': list_barcodes,
     'user_barcodes': list_user_barcode,
     'Description': list_desc,
     'Brand_code':list_brandcode,
     'rewards_code': list_rewards_group,
     'reward_partner_id':reward_partner_id,
     'points_payer_id': points_payer_id
     
    })
#concatenating the above created dataframe to existing dataframe

df_receipts_new= pd.concat([df_receipts, barcode_df], axis=1)
```

```python
df_receipts_new=df_receipts_new.replace('Not Available',np.nan)
```
